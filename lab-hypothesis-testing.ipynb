```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy import stats

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
df_dragon = df[(df["Type 1"]=="Dragon") | (df["Type 2"]== "Dragon")]
df_grass = df[(df["Type 1"]=="Grass") | (df["Type 2"]== "Grass")]

alpha = 0.05

# Calculate dragon statistics
dragon_stat = df_dragon['HP'].agg(
    mean='mean',
    median='median',
    variance='var',
    std_dev='std',
    sample_size="count"
)

# Calculate grass statistics
grass_stat = df_grass['HP'].agg(
    mean='mean',
    median='median',
    variance='var',
    std_dev='std',
    sample_size="count"
)

print("Dragon statistics:")
display(dragon_stat)
print("Grass statistics:")
display(grass_stat)

# Hypothesis: 
#H1: mu_price dragon > mu_price grass
#H0: mu_price dragon <= mu_price grass

dragon_stat_hp = df_dragon["HP"]
grass_stat_hp = df_grass["HP"]

stat, p_val =st.ttest_ind(dragon_stat_hp,grass_stat_hp, equal_var=False, alternative = "greater")

if p_val > alpha:
    print(f"Since {round(p_val, 5)} > {alpha}, we cannot reject the null hypothesis. "
          "This suggests that there is not enough evidence to say Dragon-type PokÃ©mon have more HP than Grass-type PokÃ©mon.")
else:
    print(f"Since {round(p_val, 5)} < {alpha}, we reject the null hypothesis. "
          "This suggests that Dragon-type PokÃ©mon have, on average, significantly more HP than Grass-type PokÃ©mon.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here
legendary_stats = df[df['Legendary'] == 1]
non_legendary_stats = df[df['Legendary'] == 0]

statistics = {}
sts = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
for stat in sts :
    f_stat, p_val = stats.f_oneway(legendary_stats[stat], non_legendary_stats[stat])
    statistics[stat] = (f_stat, p_val)

alpha = 0.05

for stat, (f_stat, p_val) in statistics.items():
    if p_val < alpha:
        print(f"Stat: {stat} has a P-value of {round(p_val,4)}. Since the P-value is lower than {alpha}, we reject H0: there is a significant difference in stats between Legendary and Non-Legendary PokÃ©mon.\n")
    else:
        print(f"Stat: {stat} has a P-value of {round(p_val,4)}. Since the P-value is higher than {alpha}, we fail to reject H0: there is no significant difference in stats between Legendary and Non-Legendary PokÃ©mon.\n")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import math as m

def euclidean(row, reference):
    """
    Input: dataframe with longitude and latitude as columns
    Input: reference as a tuple (x,y)
    """
    long = row["longitude"] # X (longitude)
    lat = row["latitude"] # Y (latitude)
    (x,y) = reference
    
    distance = m.sqrt((long - x)**2 + (lat - y)**2)
    
    return distance

school = (-118, 37)
hospital = (-122, 34)

df["distance_hospital"] = df.apply(euclidean, axis=1, reference=hospital)
df["distance_school"] = df.apply(euclidean, axis=1, reference=school)

thresh = 0.5
df["close_hospital"] = df.apply(lambda x: "close" if x["distance_hospital"] < thresh else "far", axis=1)
df["close_school"] = df.apply(lambda x: "close" if x["distance_school"] < thresh else "far", axis=1)

df_close = df[(df["close_hospital"] == "close" ) | (df["close_school"] == "close" )]["median_house_value"]
df_far = df[(df["close_hospital"] == "far" ) & (df["close_school"] == "far" )]["median_house_value"]

# Calculate statistics
df_close_stat = df_close.agg(
    mean='mean',
    median='median',
    variance='var',
    std_dev='std',
    sample_size="count"
)

df_far_stat = df_far.agg(
    mean='mean',
    median='median',
    variance='var',
    std_dev='std',
    sample_size="count"
)

# Display statistics
print("Statistics for houses close to school/hospital:")
print(df_close_stat)
print("\nStatistics for houses far from school/hospital:")
print(df_far_stat)

#Hypothesis:
# H1: Houses close to either a school or a hospital are more expensive
# H0: Houses close to either hospital or school are not more expensive


stat, p_val = st.ttest_ind(df_close,df_far, equal_var=False, alternative="greater")
alpha = 0.05
if p_val > alpha:
    print(f"P-value of {p_val} > {alpha} (significance): we can't reject H0, therefore we can't affirm that the houses are more expensive.")
elif p_val < alpha:
    print(f"P-value of {p_val} < {alpha} (significance): we reject H0, so the houses near the hospital or school are indeed more expensive.")
```
